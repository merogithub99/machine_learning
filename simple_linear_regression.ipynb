```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
dataset =pd.read_csv("cgpa_salary_dataset.csv")
```

```python
dataset.head()
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x="CGPA",y="Salary_Package(LPA)",data=dataset)
plt.show()
```

```python
x=dataset[["CGPA"]]
y=dataset["Salary_Package(LPA)"]
x.ndim
# x should be two dimensional
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
# i.e 20 percentage
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.score(x_test,y_test) 
# ie.accuracy score 
# change random state to increase score
```

```python
lr.fit(x_train,y_train)
```

```python
dataset.head(1)
```

```python
lr.predict([[6.87]])
```

```python
lr.predict([[9.87]])
```

```python
# eqn ----> y=mx +c
m=lr.coef_
m
```

```python
c=lr.intercept_
c
```

```python
# y=mx+c
m*9.87 + c
```

same predicted value using lr.predict for 9.87 gpa and value of y

```python
y_precict=lr.predict(x)
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x="CGPA",y="Salary_Package(LPA)",data=dataset)
plt.plot(dataset["CGPA"],y_precict,c="red")
plt.legend(["org data","predict line "])
plt.savefig("precit.jpg")
```
