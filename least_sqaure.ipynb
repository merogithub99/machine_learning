```python
import numpy as np
import matplotlib.pyplot as plt

# Fake simple data
x = np.array([1, 2, 3, 4, 5])
y = np.array([1.5, 2.8, 3.2, 4.5, 5.1])

# Mean values
x_mean = np.mean(x)
y_mean = np.mean(y)

# Calculating slope (m)
m = np.sum((x - x_mean) * (y - y_mean)) / np.sum((x - x_mean)**2)

# Calculating intercept (c)
c = y_mean - m * x_mean

print("Slope m =", m)
print("Intercept c =", c)

# Prediction line
y_pred = m * x + c

# Plot
plt.scatter(x, y)
plt.plot(x,y_pred)
plt.show()
```

⭐ What this code is doing

✔ Step 1: Compute mean of x and y

✔ Step 2: Use least squares formula to find slope & intercept

✔ Step 3: Predict values

✔ Step 4: Plot actual points vs predicted line
