```python
import pandas as pd
import numpy as np 
```

Every time you run the code, NumPy generates new random numbers.
But by setting a seed value, the random numbers remain the same on every run.

```python
np.random.seed(42)
```

Picks random floating numbers
Range: 0.0 to 20.0
Total: 1000 values
Rounded to 1 decimal place

```python
# age between 20-60
age=np.random.randint(20,60,1000) 
experience = np.round(np.random.uniform(0,20,1000),1)
# salary formula with randomness
```

```python
salary = np.round(20000 + (experience * 5000) + (age * 300) + np.random.normal(0, 5000, 1000), 2)
```

```python
df=pd.DataFrame({
  "Age":age,
  "Experience":experience,
  "Salary":salary
})
```

```python
df.head()
```

```python
file_path="age_exp_salary.csv"
df.to_csv(file_path,index=False)
# index=False avoids adding an extra index column
```
